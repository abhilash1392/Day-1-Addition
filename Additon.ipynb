```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
## creating a 3 list of 10000 numbers from 0 to 10000 different combinations randomly

list_1 = np.random.normal(0,10,size=(10000))
list_2 = np.random.normal(2,100,size=(10000))
list_3 = np.random.normal(3,1000,size=(10000))
```

```python
plt.plot(list_1)
plt.plot(list_2)
plt.plot(list_3)
```

```python
## Creating a dataframe with three list

df = pd.DataFrame(zip(list_1,list_2,list_3),columns=['x1','x2','x3'])
```

```python
df.head()
```

```python
df['y'] = df['x1']+df['x2']+df['x3']
```

```python
df.head()
```

```python
plt.plot(df['y'].values)
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
train,test = train_test_split(df,test_size=0.2,random_state=42) 
```

```python
train.shape
```

```python
test.shape
```

```python
x_train = train.drop(['y'],axis=1).values
y_train = train['y'].values
x_test = test.drop(['y'],axis=1).values
y_test = test['y'].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error,mean_squared_error
```

```python
mean_absolute_percentage_error(y_true=y_test,y_pred=y_pred)
```

```python
mean_squared_error(y_true=y_test,y_pred=y_pred)
```
